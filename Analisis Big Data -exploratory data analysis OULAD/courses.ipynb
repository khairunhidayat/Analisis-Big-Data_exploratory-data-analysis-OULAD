```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Menghubungkan antara colab dengan akun gooogle drive penyimpanan file
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Fungsi ini mengambil satu argumen, yaitu path (lokasi) file CSV yang ingin dibaca.
df = pd.read_csv('/content/drive/MyDrive/Analisis Big Data/courses.csv')
df
```

```python
# Melihat informasi dasar tentang dataset
print("Informasi Dataset:")
print(df.info())

# Menampilkan 5 baris pertama dari dataset
print("\n5 Baris Pertama Dataset:")
print(df.head())

# Statistik deskriptif untuk kolom numerik
print("\nStatistik Deskriptif:")
print(df.describe())

# Melihat jumlah missing values pada setiap kolom
print("\nJumlah Missing Values pada Setiap Kolom:")
print(df.isnull().sum())
```

```python
# Menampilkan jumlah missing values pada setiap kolom
missing_data = df.isnull().sum()

# Menampilkan kolom-kolom dengan missing values (jika ada)
columns_with_missing_data = missing_data[missing_data > 0]
print("Kolom dengan Missing Values:")
print(columns_with_missing_data)

# Menampilkan jumlah total missing values dalam dataset
total_missing_values = missing_data.sum()
print("\nTotal Missing Values dalam Dataset:", total_missing_values)
```

```python
# Visualisasi distribusi data untuk kolom 'code_module'
plt.figure(figsize=(10, 6))
sns.histplot(df['code_module'], bins=10, kde=True)
plt.title('Distribusi code_module')
plt.xlabel('code_module')
plt.ylabel('module_presentation_length')
plt.show()
```

setelah melakukan Visualisasi distribusi data untuk kolom Code Modul didapatkan hasil yaitu garis bar dari awal sudah mulai naik hingga titik tertinggi berada di FFF dan selanjutnya turun

```python
# Visualisasi perhitungan data dari code presentation
plt.figure(figsize=(10, 6))
sns.countplot(x='code_presentation', data=df)
plt.xlabel('Code Presentation')
plt.ylabel('Count')
plt.title('Perhitungan Data Dari Code Presentation')
plt.xticks(rotation=45)
plt.show()
```
