```python
#Fungsi ini mengambil satu argumen, yaitu path (lokasi) file CSV yang ingin dibaca.
df = pd.read_csv('/content/drive/MyDrive/Analisis Big Data/studentInfo.csv')
df
```

```python
# Melihat informasi dasar tentang dataset
print("Informasi Dataset:")
print(df.info())

# Menampilkan 5 baris pertama dari dataset
print("\n5 Baris Pertama Dataset:")
print(df.head())

# Statistik deskriptif untuk kolom numerik
print("\nStatistik Deskriptif:")
print(df.describe())

# Melihat jumlah missing values pada setiap kolom
print("\nJumlah Missing Values pada Setiap Kolom:")
print(df.isnull().sum())
```

```python
# Menampilkan jumlah missing values pada setiap kolom
missing_data = df.isnull().sum()

# Menampilkan kolom-kolom dengan missing values (jika ada)
columns_with_missing_data = missing_data[missing_data > 0]
print("Kolom dengan Missing Values:")
print(columns_with_missing_data)

# Menampilkan jumlah total missing values dalam dataset
total_missing_values = missing_data.sum()
print("\nTotal Missing Values dalam Dataset:", total_missing_values)
```

```python
# untuk mengisi nilai-nilai yang hilang (missing values) dalam kolom 'imd_band' dari dataframe df dengan nilai 0.
# digunakan untuk mengisi nilai-nilai yang hilang dalam suatu kolom dengan nilai tertentu, dalam hal ini, nilai 0.
df['imd_band'].fillna(0, inplace=True)
```

```python
# Menampilkan jumlah missing values pada setiap kolom
missing_data = df.isnull().sum()

# Menampilkan kolom-kolom dengan missing values (jika ada)
columns_with_missing_data = missing_data[missing_data > 0]
print("Kolom dengan Missing Values:")
print(columns_with_missing_data)

# Menampilkan jumlah total missing values dalam dataset
total_missing_values = missing_data.sum()
print("\nTotal Missing Values dalam Dataset:", total_missing_values)
```

```python
# Visualisasi distribusi data untuk kolom 'id_assessment'
sns.countplot(x='gender', data=df)
plt.title('Distribusi Gender')
plt.show()

sns.countplot(x='final_result', data=df)
plt.title('Distribusi Hasil Akhir')
plt.show()
```

dari visual Distribusi Gender diatas didapatkan perhitungan gender male lebih banyak dari pada female, kemudian untuk distribusi hasil akhir didapatkan bar Pass lebih banyak dari pada bar lainnya
